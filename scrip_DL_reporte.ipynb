```python
import os
import pandas as pd
```

```python
# Ruta de la carpeta con los archivos CSV
archivo = r"C:\Users\ERBC\Desktop\python\Reporte_DL\Análisis - Delivery late W02 - Análisis.csv"
```

```python
# Lista de columnas requeridas
columnas_deseadas = [
    "Agente", "CASOS ERROR",'Si aplica, ¿Que error crítico de usuario comete? ', 'Si aplica, ¿Que error crítico de usuario comete? .1', 
    'Si aplica, ¿Que error crítico de negocio comete? ', 'Si aplica, ¿Que error crítico de negocio comete? .1', 'Si aplica, ¿Que error no crítico comete? ', 
    'Si aplica, ¿Que error no crítico comete? .1', 'Si aplica, ¿Cuál es el error? 1', 'Si aplica, ¿Cuál es el error? 2', 
    'Si aplica ¿Comete error crítico de usuario en comunicación? ', 
    'Si aplica ¿Comete error crítico de usuario en comunicación? .1', 'Si aplica ¿Comete error crítico de negocio en comunicación? ', 'Si aplica ¿Comete error no crítico  en comunicación? ', 
    'Si aplica ¿Comete error no crítico  en comunicación? .1' 

]
# Lista para almacenar los DataFrames procesados
dataframes_filtrados = []

df = pd.read_csv(archivo)
df_filtrado = df[columnas_deseadas]
df_filtrado
```

```python
df_filtrado = df_filtrado.astype(str)
# Eliminar espacios en blanco o caracteres no visibles en los nombres de agentes
df_filtrado['Agente'] = df_filtrado['Agente'].str.strip()

# Obtener la lista de agentes únicos después de limpiar
agentes = sorted(df_filtrado['Agente'].dropna().unique())
# Diccionario para almacenar los resultados
conteos = {}

agentes
```

```python
# Iterar sobre los agentes y contar los valores en las demás columnas
for agente in agentes:
    # Filtrar filas para el agente actual
    datos_agente = df_filtrado[df_filtrado['Agente'] == agente]
    
    # Eliminar las filas que contengan "Sí" o "No" en todas las columnas excepto "Agente"
    datos_agente = datos_agente.replace({'Si': None, 'No': None, "nan":None})
    
    # Contar valores únicos en las demás columnas (excluyendo 'Agente')
    conteos[agente] = (
        datos_agente.drop(columns='Agente')
        .apply(pd.Series.value_counts)  # Contar valores
        .sum(axis=1)  # Sumar por fila
        .to_dict()
    )

# Mostrar el resultado ordenado
for agente, conteo in conteos.items():
    print(f"{agente}: {conteo}")
```

```python
# Convertir el diccionario de conteos a un DataFrame
df_conteos = pd.DataFrame.from_dict(conteos, orient='index').fillna(0)

# Ordenar el DataFrame alfabéticamente por el nombre del agente (índice)
df_conteos = df_conteos.sort_index(ascending=True)

# Eliminar los decimales convirtiendo los valores a enteros
df_conteos = df_conteos.astype(int)

# Mostrar el DataFrame resultante
print(df_conteos)
df_conteos
```

```python
# Obtener la fecha actual
fecha_actual = pd.to_datetime("today")

# Obtener el número de la semana actual (y restar 1 para obtener la semana anterior)
numero_semana_actual = fecha_actual.isocalendar()[1]
numero_semana_anterior = numero_semana_actual - 1
```

```python
# Si quieres exportarlo a CSV
df_conteos.to_csv(f'conteos_por_agente_W_{numero_semana_anterior}.csv', index=True)
```

```python
# Crear un archivo de texto para guardar los resultados
with open(f'errores_agentes_W_ {numero_semana_anterior}.txt', 'w') as f:
    # Iterar sobre las filas del DataFrame (por cada agente)
    for agente, conteos in df_conteos.iterrows():
        # Escribir el nombre del agente
        nombre_usuario = agente.split('.ext')[0]
        f.write(f"Errores para agente: {agente} @{nombre_usuario}\n")
        f.write(f"Para la Week{numero_semana_anterior}\n")
        
        # Ordenar los errores alfabéticamente antes de iterar
        for error, cantidad in sorted(conteos.items()):
            if cantidad > 0:  # Solo escribir si la cantidad es mayor que 0
                f.write(f"  - {error}: {cantidad}\n")
        f.write("Para asegurar un servicio excelente a PedidosYa, debemos priorizar corregir estos errores,\nmostrando nuestro compromiso con la mejora continua y la satisfacción del cliente.\n\n")
        f.write("Por favor reaccionar a este mensaje o comentar para validar que se ha recibido la información.\n\n")
        # Agregar una línea en blanco después de cada agente
        f.write("\n")

# Confirmar que el archivo ha sido creado
print("El archivo de errores ha sido generado exitosamente.")
```
