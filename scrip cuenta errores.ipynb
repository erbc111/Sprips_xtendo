```python
import os
import pandas as pd

# Ruta de la carpeta con los archivos CSV
carpeta = r"C:\Users\ERBC\Desktop\python\ds"

# Lista de columnas requeridas
columnas_deseadas = [
    "#", "agent", "team_name", "evaluated_by", "etiqueta_auditada",
    "section_1_answer_2_feedback_error_1_ec_u", "section_1_answer_2_feedback_error_2_ec_u",
    "section_1_answer_2_feedback_error_3_ec_u", "section_1_answer_3_feedback_error_1_ec_n",
    "section_1_answer_3_feedback_error_1_ec_n_que_etiqueta_correspondia",
    "section_1_answer_3_feedback_error_2_ec_n", "section_1_answer_3_feedback_error_3_ec_n",
    "section_1_answer_4_feedback_error_1_enc", "section_1_answer_4_feedback_error_2_enc",
    "section_1_answer_4_feedback_error_3_enc", "section_2_answer_1_feedback_ecu_1_comunicacion",
    "section_2_answer_1_feedback_ecu_1_comunicacion_informacion_incompleta",
    "section_2_answer_1_feedback_ecu_2_comunicacion",
    "section_2_answer_1_feedback_ecu_2_comunicacion_informacion_incompleta_incorrecta",
    "section_2_answer_1_feedback_ecu_3_comunicacion", "section_2_answer_2_feedback_ecn_comunicacion_1",
    "section_2_answer_2_feedback_ecn_comunicacion_2", "section_2_answer_3_feedback_enc_comunicacion_1",
    "section_2_answer_3_feedback_enc_comunicacion_2", "section_2_answer_3_feedback_enc_comunicacion_3"
]

# Lista para almacenar los DataFrames procesados
dataframes_filtrados = []

# Iterar sobre los archivos CSV
for archivo in os.listdir(carpeta):
    if archivo.endswith(".csv"):
        ruta_completa = os.path.join(carpeta, archivo)
        # Cargar el archivo CSV
        df = pd.read_csv(ruta_completa)
        # Filtrar solo las columnas deseadas que existen en el archivo
        columnas_presentes = [col for col in columnas_deseadas if col in df.columns]
        df_filtrado = df[columnas_presentes]
        dataframes_filtrados.append(df_filtrado)

# Concatenar todos los DataFrames filtrados en uno solo
df_unido = pd.concat(dataframes_filtrados, ignore_index=True)

# Guardar el DataFrame final en un archivo CSV (opcional)
df_unido.to_csv("dataframe_unido.csv", index=False)

print("Unión completada. El DataFrame final tiene las siguientes columnas:")
print(df_unido.columns)
```
